```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
from tabulate import tabulate 
import warnings 
import textwrap

warnings.filterwarnings("ignore") 

plt.rcdefaults()

figw, figh = plt.rcParams["figure.figsize"]

sns.set()
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)

# μέγεθος δικτύου: Τελευταίο ψηφίο ΑΜ = 5
n = 45
d = 4
p = 0

# Make figure much larger so all 4 connections from each node are visible
size = 40
plt.figure(figsize=(size, size))

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n, d, p)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

**Σημείωση:** Στην παραπάνω εικόνα κάθε κόμβος είναι συνδεδεμένος με 4 άλλους κόμβους (τους 2 αμέσως δεξιά του και τους 2 αμέσως αριστερά του). Αυτό μπορεί να γίνει πιο εμφανές με μεγέθυνση στην εικόνα.

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M) 

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)

len(dict(RGER.degree()).values())
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
d = 4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n, d) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
d = 4
p = 0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n, d, p) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

## B. Μετρικές δικτύου

```python
# Graphs with Names
GN = [(REG, "REG"), (RGER, "RGER"), (RGG, "RGG"), (SF, "SF"), (SW, "SW")]
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):

#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.

```python
def cc(G):
    return list(nx.clustering(G).values())

def plot_distrib(func, title, suptitle=""):
    fig, ax = plt.subplots(2, 3, figsize=(2*figw, 1.5*figh))

    for i, (G, name) in enumerate(GN):    
        data = func(G)
        
        I, J = i // 3, i % 3
        
        ax[I][J].set_title(name)
        ax[I][J].set_xlabel(title)

        # set 10 different bins for REG case so we don't have one bar taking all the horizontal space
        bins = "auto" if name != "REG" else np.linspace(0.9*min(data), 1.1*max(data), 10)

        sns.histplot(data, ax=ax[I][J], bins=bins, kde=True)

        y_min = 0
        y_max = min(1.1 * n, ax[I][J].get_ylim()[1])
        
        ax[I][J].set_ylim([y_min, y_max])

    ax[1][2].set_visible(False)

    fig.suptitle(suptitle)

    fig.tight_layout()
    
plot_distrib(cc, "Clustering coefficient", "Clustering coefficient distribution")
```

#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.

```python
table = []
for G, name in GN:
	table.append([name, np.mean(cc(G)), np.var(cc(G))])

print(tabulate(table, headers=["Type", "Average clustering coefficient", "Variance of clustering coefficient"], tablefmt="fancy_grid"))
```

#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 

```python
table = []
for G, name in GN:
	minimum = np.min(cc(G))
	maximum = np.max(cc(G))

	nodes_with_min = [str(node) for (node, cc) in nx.clustering(G).items() if cc == minimum]
	nodes_with_max = [str(node) for (node, cc) in nx.clustering(G).items() if cc == maximum]

	nodes_with_min = ", ".join(nodes_with_min)
	nodes_with_max = ", ".join(nodes_with_max)

	table.append([name, textwrap.fill(nodes_with_min, 30), minimum, textwrap.fill(nodes_with_max, 30), maximum])

print(tabulate(table, headers=["Type", "Nodes with min clustering coefficient", "Min", "Nodes with max clustering coefficient", "Max"], tablefmt="fancy_grid"))
```

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
table = []
used_name = {name: False for _, name in GN}

for G, name in GN:
    diameter_paths = {node: list(paths.values()).count(nx.diameter(G)) for node, paths in nx.all_pairs_shortest_path_length(G)}
    
    maximum = max(diameter_paths.values())

    for node, paths in nx.all_pairs_shortest_path_length(G):
        if diameter_paths[node] == maximum:
            if used_name[name]:
                name_for_table = ""
                maximum_for_table = ""
            else:
                used_name[name] = True
                name_for_table = name
                maximum_for_table = maximum

            table.append([name_for_table, maximum_for_table, node, [node for node, length in paths.items() if length == nx.diameter(G)]])

table_headers = [
    "Type", 
    textwrap.fill("Maximum number of paths equal to diameter", 20), 
    textwrap.fill("Node with maximum number of paths", 20),
    "To nodes"
]
print(tabulate(table, headers=table_headers, tablefmt="fancy_grid"))
```

#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της.

```python
def find_periphery(G):
    return nx.periphery(G)

def find_center(G):
    return nx.center(G)

table = []
for G, name in GN:
    periphery_for_table = textwrap.fill(", ".join([str(item) for item in find_periphery(G)]), 30)
    center_for_table = textwrap.fill(", ".join([str(item) for item in find_center(G)]), 30)

    table.append([name, periphery_for_table, center_for_table])
print(tabulate(table, headers=["Type", "Periphery", "Center"], tablefmt="fancy_grid"))
```

### Γ4. Kεντρικότητα Katz

##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality, να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ίση με $\frac{1}{λ_{max}}-0.01$.

```python
def find_katz(G):
    l_max = nx.adjacency_spectrum(G).max()
    alpha = np.real(1/l_max - 0.01)
    return list(nx.katz_centrality(G, alpha).values())

plot_distrib(find_katz, "Katz centrality", "Katz centrality distribution")
```

##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.

```python
table = []
for G, name in GN:
	table.append([name, np.mean(find_katz(G)), np.var(find_katz(G))])

print(tabulate(table, headers=["Type", "Average Katz centrality", "Variance of Katz centrality"], tablefmt="fancy_grid"))
```

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. 

```python
def degree_dict(G):
    return dict(G.degree())

def closeness_dict(G):
    return nx.closeness_centrality(G)

def betweenness_dict(G):
    return nx.betweenness_centrality(G)

def katz_dict(G):
    l_max = nx.adjacency_spectrum(G).max()
    alpha = np.real(1/l_max - 0.01)
    return nx.katz_centrality(G, alpha)

def find_central(G):
    central_degree = [node for node, degree in degree_dict(G).items() if degree == max(find_degree(G))]
    central_closeness = [node for node, closeness in closeness_dict(G).items() if closeness == max(find_closeness(G))]
    central_betweenness = [node for node, betweenness in betweenness_dict(G).items() if betweenness == max(find_betweenness(G))]
    central_katz = [node for node, katz in katz_dict(G).items() if katz == max(find_katz(G))]

    return central_degree, central_closeness, central_betweenness, central_katz

table = []
for G, name in GN:
    central_degree, central_closeness, central_betweenness, central_katz = find_central(G)

    wrap_amount = 15

    table_line = [
        name, 
        textwrap.fill(", ".join([str(item) for item in central_degree]), wrap_amount), 
        max(find_degree(G)),
        textwrap.fill(", ".join([str(item) for item in central_closeness]), wrap_amount),
        max(find_closeness(G)),
        textwrap.fill(", ".join([str(item) for item in central_betweenness]), wrap_amount),
        max(find_betweenness(G)),
        textwrap.fill(", ".join([str(item) for item in central_katz]), wrap_amount),
        max(find_katz(G))
    ]
    table.append(table_line)

print(tabulate(table, headers=["Type", "By degree", "Max degree", "By closeness", "Max closeness", "By betweenness", "Max betweenness", "By Katz", "Max Katz"], tablefmt="fancy_grid"))
```

#### Να σχολιαστούν τα αποτελέσματα.

Εδώ εφαρμόζουμε τις παραπάνω κεντρικότητες προκειμένου να διαπιστώσουμε ποιοι κόμβοι είναι πιο κομβικοί (pun intended) για το δίκτυο. Παρατηρούμε ότι κάποιοι κόμβοι είναι οι πιο κεντρικοί με βάση παραπάνω από μία κεντρικότητες. Παραδείγματα είναι:

* ακραία περίπτωση είναι ο κόμβος 15 στην τοπολογία RGG, που είναι ο πιο κεντρικός με βάση και τις 4 κεντρικότητες

Στα παραπάνω δίκτυα μπορούμε λοιπόν να εξάγουμε πληροφορία για την σημασία των κεντρικών κόμβων, και να πάρουμε σχεδιαστικές αποφάσεις με βάση τα δεδομένα αυτά.
```python
# Connectivity percentage function
def connper(creator, kwargs):
	num_conn = 0
	for i in range(100):
		G = creator(**kwargs)
		num_conn += 1 if nx.is_connected(G) else 0
	return num_conn / 100

# NOTE: All following ranges are half-open e.g. range(2, 12, 2) means [2, 4, 6, 8, 10], 12 NOT included

# Connectivity REG
random.seed(seedno)
conn_REG = []
for n in [100, 200]:
	for d in range(2, 12, 2):
		percent = connper(nx.watts_strogatz_graph, {"n": n, "k": d, "p": 0})

		conn_REG.append({"n": n, "d": d, "percent": percent})

# Connectivity RG Erdos-Renyi
random.seed(seedno)
conn_RG_ER = []
for n in [100, 200]:
	for M in range(100, 900, 100):
		percent = connper(nx.gnm_random_graph, {"n": n, "m": M})

		conn_RG_ER.append({"n": n, "M": M, "percent": percent})

# Connectivity RG Gilbert
random.seed(seedno)
conn_RG_G = []
for n in [100, 200]:
	for p in np.arange(0.01, 0.11, 0.01):
		percent = connper(nx.gnp_random_graph, {"n": n, "p": p})

		conn_RG_G.append({"n": n, "p": p, "percent": percent})

# Connectivity RGG
random.seed(seedno)
conn_RGG = []
for n in [100, 200]:
	for R in np.arange(0.025, 0.275, 0.025):
		percent = connper(nx.random_geometric_graph, {"n": n, "radius": R})

		conn_RGG.append({"n": n, "R": R, "percent": percent})

# Connectivity SF
random.seed(seedno)
conn_SF = []
for n in [100, 200]:
	for d in range(2, 12, 2):
		percent = connper(nx.barabasi_albert_graph, {"n": n, "m": d})

		conn_SF.append({"n": n, "d": d, "percent": percent})

# Connectivity SW
random.seed(seedno)
conn_SW = []
for n in [100, 200]:
	for d in range(2, 12, 2):
		for g_p in np.arange(0.1, 0.8, 0.1):
			percent = connper(nx.watts_strogatz_graph, {"n": n, "k": d, "p": g_p})

			conn_SW.append({"n": n, "d": d, "g_p": g_p, "percent": percent})
```

Διαγράμματα

```python
fig, ax = plt.subplots(3, 3, figsize=(2.2*figw, 2.5*figh))

# Convert data from list of dictionaries (lod) to dictionary of lists (dol) for plotting
def lod2dol(lod):
	return {key: [dictionary[key] for dictionary in lod] for key in lod[0]}

# Plot REG
currax = ax[0][0]
sns.lineplot(data=lod2dol(conn_REG), x="d", y="percent", hue="n", palette=["tab:red", "tab:green"], ax=currax)
currax.set_title("Connectivity REG")

# Plot RG Erdos-Renyi
currax = ax[0][1]
sns.lineplot(data=lod2dol(conn_RG_ER), x="M", y="percent", hue="n", palette=["tab:red", "tab:green"], ax=currax)
currax.set_title("Connectivity RG Erdos-Renyi")

# Plot RG Gilbert
currax = ax[0][2]
sns.lineplot(data=lod2dol(conn_RG_G), x="p", y="percent", hue="n", palette=["tab:red", "tab:green"], ax=currax)
currax.set_title("Connectivity RG Gilbert")

# Plot RGG
currax = ax[1][0]
sns.lineplot(data=lod2dol(conn_RGG), x="R", y="percent", hue="n", palette=["tab:red", "tab:green"], ax=currax)
currax.set_title("Connectivity RGG")

# Plot SF
currax = ax[1][1]
sns.lineplot(data=lod2dol(conn_SF), x="d", y="percent", hue="n", palette=["tab:red", "tab:green"], ax=currax)
currax.set_title("Connectivity SF")

# Plot SW

for n in [100, 200]:
	currax = ax[2][n // 100 - 1]
	
	# Filter to get values corresponding to n = 100, n = 200 respectively
	filtered = list(filter(lambda d: d["n"] == n, conn_SW))
	
	data = []
	for d in range(2, 12, 2):
		line = []
		for g_p in np.arange(0.1, 0.8, 0.1):
			for dictionary in filtered:
				if dictionary["d"] == d and dictionary["g_p"] == g_p:
					line.append(dictionary["percent"])
		data.append(line)
					
	sns.heatmap(
		data, 
		ax=currax, 
		cbar_kws={"label": "percentage"}, 
		cmap="copper",
		xticklabels=np.round(np.arange(0.1, 0.8, 0.1), 2),
		yticklabels=range(2, 12, 2)
	).set(xlabel="$g_p$", ylabel="d")

	currax.set_title(f"Connectivity SW, n = {n}")

for i, j in [(1, 2), (2, 2)]:
	ax[i][j].set_visible(False)

fig.suptitle("Connectivity\n\n")

fig.tight_layout()
```

### E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$ της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ (αντικαταστήστε το $x$ με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. 

```python
random.seed(seedno)
n = 305
d = 4

results = []
for g_p in np.arange(0, 1.1, 0.1):
    G = nx.watts_strogatz_graph(n, d, g_p)
    avg_path_length = np.mean(compute_path_lengths(G))
    avg_cluster_coeff = np.mean(cc(G))
    results.append({"g_p": g_p, "avg_path_length": avg_path_length, "avg_cluster_coeff": avg_cluster_coeff})

fig, ax = plt.subplots(1, 2, figsize=(1.75*figw, figh))

lineplotax = sns.lineplot(data=lod2dol(results), x="g_p", y="avg_path_length", ax=ax[0])
lineplotax.set_xlabel("$g_p$")
lineplotax.set_ylabel("Average path length")
lineplotax.set_title("Average path length")

lineplotax = sns.lineplot(data=lod2dol(results), x="g_p", y="avg_cluster_coeff", ax=ax[1])
lineplotax.set_xlabel("$g_p$")
lineplotax.set_ylabel("Average clustering coefficient")
lineplotax.set_title("Average clustering coefficient")

fig.tight_layout()
```

##### Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

Για $g_p = 0$ έχουμε πλέγμα, και άρα το μέσο μήκος μονοπατιού καθώς και ο συντελεστής ομαδοποίησης είναι μεγάλα.

Για $g_p = 1$ έχουμε τυχαίο γράφο, και άρα το μέσο μήκος μονοπατιού μικραίνει, ενώ ο συντελεστής ομαδοποίησης σχεδόν μηδενίζεται.

Συγκεκριμένα, για τιμές $g_p = 0-0.2$ το μέσο μήκος μονοπατιού μειώνεται σχεδόν κατακόρυφα, ενώ για τις υπόλοιπες τιμές $g_p = 0.3-1$ το μέσο μήκος μονοπατιού σταθεροποιείται στην τιμή $5$.

Αντίστοιχα, για τιμές $g_p = 0-0.5$ έχουμε σταδιακή πτώση του clustering coefficient, και για $g_p = 0.6-1$ το clustering coefficient γίνεται σχεδόν μηδέν.

Θα μπορούσαμε λοιπόν να πούμε ότι για $g_p = 0.1-0.5$ το δίκτυο συμπεριφέρεται σαν δίκτυο SW.

Απεικονίζουμε τις κατανομές βαθμού και συντελεστή ομαδοποίησης, αλλά και το ίδιο το δίκτυο προκειμένου να αποκτήσουμε μία εικόνα για τη δομή του, και να την εντάξουμε σε κάποιον από τους "τεχνητούς" τύπους δικτύων που μελετήσαμε ως τώρα. Στην απεικόνιση του δικτύου, το μέγεθος κάθε κόμβου είναι ευθέως ανάλογο των συνδέσεων που έχει συνάψει. 

Η πρώτη σημαντική ένδειξη για τον τύπο του δικτύου εμφανίζεται με την κατανομή του βαθμού των κόμβων, η οποία είναι φυσικά κατανομή Νόμου Δύναμης, όπου λίγοι χαρακτήρες έχουν εκτελέσει πολλές αλληλεπιδράσεις, και οι υπόλοιποι είναι λιγότερο σημαντικοί και έχουν πραγματοποιήσει λιγότερες αλληλεπιδράσεις.

Επιπλέον, βλέπουμε ότι ένα μεγάλο ποσοστό των κόμβων έχει υψηλό clustering coefficient, και επομένως ο μέσος όρος του clustering coefficient είναι επίσης μεγάλος.

Τέλος, ελέγχοντας το μέσο μήκος μονοπατιού, βλέπουμε ότι είναι επίσης μικρό, και κάθε χαρακτήρας συνδέεται με λιγότερα από 3 hops με κάθε έναν άλλον κατά μέσο όρο. 

Λαμβάνοντας υπόψιν τα παραπάνω χαρακτηριστικά, αλλά και συγκρίνοντας τις μετρικές της τοπολογίας του GoT με τις αντίστοιχες των τεχνητών δικτύων, μπορούμε να πούμε με σιγουριά ότι το δίκτυο της πρώτης σεζόν του GoT είναι ***Scale Free***.
